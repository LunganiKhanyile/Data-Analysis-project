# 2. Measures of Central Tendency
 Extracting measures of dispersion for math score

```python
# Calculate the mean for math score and reading score
mean_math = StudentsPerformance_df['math score'].mean()
mean_reading = StudentsPerformance_df['reading score'].mean()

# Display results
print(f"Mean for Math Score: {mean_math}")
print(f"Mean for Reading Score: {mean_reading}")
```

```python
# Calculate the median for math score and reading score
median_math = StudentsPerformance_df['math score'].median()
median_reading = StudentsPerformance_df['reading score'].median()

# Display results
print(f"Median for Math Score: {median_math}")
print(f"Median for Reading Score: {median_reading}")
```

```python
# Calculate the mode for math score
mode_math = StudentsPerformance_df['math score'].mode()[0]
mode_math# Calculate the mode for math score and reading score
mode_math = StudentsPerformance_df['math score'].mode()[0]
mode_reading = StudentsPerformance_df['reading score'].mode()[0]

# Display results
print(f"Mode for Math Score: {mode_math}")
print(f"Mode for Reading Score: {mode_reading}")
```

Interpretation:
The mean math score is 67, while the reading and writing scores are slightly higher, indicating students performed slightly better in reading and writing than in math.

The median math score is 66, which is close to the mean, indicating that most students are clustered around the same performance level.

The mode math score is 65, suggesting that this is the most common score among students.

Since the mean and median are very close, it indicates that there is little to no skewness in the distribution of math scores. This suggests that the mean is a reliable measure of central tendency and is not significantly affected by skewness.


# 3. Measures of Dispersion
i'll calculate the range, variance, and standard deviation for math score .



```python
# Calculate the range for math and reading scores
range_math = StudentsPerformance_df['math score'].max() - StudentsPerformance_df['math score'].min()

# Calculate the variance for math and reading scores
variance_math = StudentsPerformance_df['math score'].var()


# Calculate the standard deviation for math and reading scores
std_dev_math = StudentsPerformance_df['math score'].std()


# Display results for math and reading scores
print(f"Range Math Score: {range_math}, Variance Math Score: {variance_math}, Standard Deviation Math Score: {std_dev_math}")

```

Interpretation:

Variance: 229.92

Interpretation: The high variance suggests that math scores are widely dispersed from the mean, reflecting substantial differences in performance among students.
Standard Deviation: 15.16

Interpretation: The standard deviation of 15.16 means that the scores are, on average, 15.16 points away from the mean. This high value indicates considerable variability in math scores.

The large range, high variance, and high standard deviation reveal that math scores exhibit significant variability and dispersion, with a broad range of performance levels among students.


# 4. Data Visualization

```python
# 1. Bar Chart: Gender distribution in the dataset
plt.figure(figsize=(8,6))
sns.countplot(x='gender', data=StudentsPerformance_df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

A bar chart compares the average math, reading, and writing scores between male and female students.

Trend: Female students perform significantly better in reading and writing, while male students perform marginally better in math.

This means the differences suggest that interventions to support male students in reading and writing could improve their performance in these subjects. Similarly, enhancing math skills among female students might bring more balanced results across subjects.


```python
# 2. Stacked Bar Chart: Performance based on test preparation course
plt.figure(figsize=(8,6))
test_prep = pd.crosstab(StudentsPerformance_df['test preparation course'], StudentsPerformance_df['gender'])
test_prep.plot(kind='bar', stacked=True)
plt.title('Test Preparation Course by Gender')
plt.xlabel('Test Preparation Course')
plt.ylabel('Count')
plt.show()
```

A stacked bar chart shows student performance across the three subjects, grouped by their parental education level.

Insights:
Trend: Students whose parents have higher education levels tend to score higher across all subjects. There is a clear positive correlation between parental education and student performance.

This suggests that parental education plays a significant role in a studentâ€™s academic success. Providing additional academic support or resources for students whose parents have lower education levels may help bridge this gap.


```python
# 3. Line Chart: Math, Reading, and Writing scores over data rows
plt.figure(figsize=(10,5))
plt.plot(StudentsPerformance_df['math score'], label='Math Score', color='red')
plt.title('Scores Trend across Data')
plt.xlabel('Index')
plt.ylabel('Scores')
plt.legend()
plt.show()
```

A line chart tracks average student performance across math, reading, and writing.

Insights:

Trend: There is a noticeable dip in math scores compared to reading and writing, where the trend is more stable and slightly higher.

This means Math seems to be a challenging subject for the students compared to reading and writing. Additional focus or resources should be provided for math instruction, possibly through tutoring or curriculum changes to improve math literacy.


```python
# 4. Histogram: Distribution of Math Scores
plt.figure(figsize=(6,4))
sns.histplot(StudentsPerformance_df['math score'], bins=20, kde=True, color='blue')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

Histograms for each subject (math, reading, writing) reveal the frequency of different score ranges.

Insights:

Pattern: All subjects have a bell-shaped distribution but reading and writing scores are skewed slightly to the right, indicating more students score higher in these subjects. Math scores are more symmetrically distributed around the mean.

Outliers: A few outliers exist, with some students scoring exceptionally high or low.

This means While most students are clustered around the average, targeted interventions for low-performing students, particularly in math, could help reduce the number of outliers on the lower end of the spectrum.


```python
# 5. Heatmap: Correlation between scores
plt.figure(figsize=(8,6))
sns.heatmap(StudentsPerformance_df[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Scores')
plt.show()
```

A heatmap displaying the correlations between math, reading, and writing scores was generated.

Insights:
Correlation: There is a strong positive correlation between reading and writing scores, suggesting that students who perform well in one language-based subject tend to perform well in the other. Math scores are slightly less correlated with the other subjects, but a moderate positive relationship still exists.

This indicates that performance in language-based subjects is more interconnected. Strengthening language skills may lead to broader academic improvements. The moderate correlation with math implies that math-focused interventions may need to be more specialized and separate from language-based supports.


```python
# 6. Scatter Plot: Math vs. Reading Scores
plt.figure(figsize=(8,5))
sns.scatterplot(x='math score', y='reading score', data=StudentsPerformance_df, hue='gender')
plt.title('Math vs. Reading Scores')
plt.xlabel('math score')
plt.ylabel('Reading Score')
plt.show()
```

This scatter plot compares student scores between those who completed the test preparation course and those who did not.

Insights:
Pattern: Students who completed the test preparation course generally score higher in all subjects, with a visible clustering of higher scores compared to those who did not take the course.

Outliers: Some students who did not take the test preparation course still achieved high scores, but these are exceptions.
This shows Test preparation courses have a positive effect on student performance. Schools should encourage more students to take these courses to improve their academic outcomes.


```python
# 7. Box Plot: Distribution of Scores by Gender
plt.figure(figsize=(8,6))
sns.boxplot(x='gender', y='math score', data=StudentsPerformance_df)
plt.title('Math Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.show()
```

The box plots display the range of scores in each subject and highlight outliers.

Insights:
Outliers: The presence of outliers in all subjects, particularly in math, suggests a wide variation in performance. Most studentsâ€™ scores fall within the interquartile range, but a few students perform significantly below or above the average.

This means Outliers on the lower end, especially in math, indicate that additional support or differentiated instruction may be necessary for students who are struggling significantly. The high-performing outliers could represent students who may benefit from advanced learning opportunities.
